```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('resnet50_training.csv')
```

```python
# Plot the different training
# summarize history for accuracy
plt.plot(df['Traning Accuracy'])
plt.plot(df['Validation Accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='lower right')
plt.show()
```

```python
# summarize history for loss
plt.plot(df['Traning NLL'])
plt.plot(df['Validation NLL'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right')
plt.show()
```
